## 4.3. GRAFICA DE MEDIDAS CENTRALES Y DISPERSION DE UNA VARAIBLE AIR TEMPERATURE

```python
df = manto_processed_df.copy()
variable = 'Air temperature [K]'
sns.histplot(
    data = df,
    x = variable
)
plt.axvline(
    x = df[variable].mean(),
    color = 'red',
    linestyle = 'dashed',
    linewidth = 2
)
plt.axvline(
    x = df[variable].median(),
    color = 'green',
    linestyle = 'dashed',
    linewidth = 2
)
plt.axvline(
    x = df[variable].median() + df[variable].std(),
    color = 'black',
    linestyle = 'dashed',
    linewidth = 2
)
plt.axvline(
    x = df[variable].quantile(0.25),
    color = 'yellow',
    linestyle = 'dashed',
    linewidth = 2
)
plt.axvline(
    x = df[variable].quantile(0.75),
    color = 'yellow',
    linestyle = 'dashed',
    linewidth = 2
)
```
